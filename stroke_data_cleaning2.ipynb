```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('healthcare-dataset-2.csv')
```

```python
# More likely to have stroke as get older
sns.histplot(data = df,x = 'age', hue = 'stroke', bins = 20)
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='stroke', y ='age', data=df)
```

```python
# status of having stroke by work type
fix, axes = plt.subplots(1, 2, figsize = (15,5))

#shows work type
sns.countplot(ax = axes[0], x = 'work_type', data = df)
#shows stroke(1) by work type
sns.countplot(ax = axes[1], x = 'work_type', hue  = 'stroke', data = df)
```

```python
# status of having stroke by smoking status
fig, axis = plt.subplots(1,2,figsize=(15,5))

sns.countplot(ax = axis[0], x = 'smoking_status', data = df)
sns.countplot(ax = axis[1], x = 'smoking_status', hue = 'stroke', data = df)
```

```python
# t-  
#   
#    
#   
#    student t,  welch's t 

#if can say the mean of specific variable in two groups divided by binary value of stroke(0 or 1) is statistically significant,
# It can be said that variable affects stroke
```

```python
from scipy import stats

data_1 = df[df['stroke'] == 1]['age']
data_0 = df[df['stroke'] == 0]['age']

#if value p is lower than 0.05, the age of two groups differs statistically significant
stats.ttest_ind(data_1, data_0)
```
